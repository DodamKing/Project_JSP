```python
import requests
from bs4 import BeautifulSoup as bs

header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Trident/7.0; rv:11.0) like Gecko'}
data = requests.get('https://www.melon.com/chart/index.htm', headers = header)
soup = bs(data.text, 'html.parser')

img = soup.select('table tr td .wrap img')
titles = soup.select('table tr td .rank01 a')
artists = soup.select('span.checkEllipsis')
```

```python
img_list = []
title_list = []
artist_list = []

for i in range(100):
    img_list.append(img[i].get('src'))
    title_list.append(titles[i].text)
    artist_list.append(artists[i].text)
```
