```python
import requests
from bs4 import BeautifulSoup as bs

data = requests.get('https://music.bugs.co.kr/musicpd/albumview/36737')
soup = bs(data.text, 'html.parser')

titles = soup.select('p.title')
artists = soup.select('p.artist')
imgs = soup.select('.list>tbody>tr')

title = []
artist = []
img = []

for i in range(100):
    img_url = imgs[i].select('a>img')
    
    title.append(titles[i].text.strip())
    artist.append(artists[i].text.split()[0])
    img.append(img_url[0].get('src'))
    
chart = []

for i in range(100):
    chart.append([title[i], artist[i], img[i]])    
```

```python
chart
```

```python
f = open('music_database.txt','a')
for i in range(100):
    f.write(chart[i][0] + ' - ' + chart[i][1] + ' - ' + chart[i][2] + '\n')
    
f.write("\n")    
f.close()
```
