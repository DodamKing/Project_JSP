```python
import requests
from bs4 import BeautifulSoup as bs

header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Trident/7.0; rv:11.0) like Gecko'}
url = 'https://www.melon.com/chart/age/index.htm?chartType=YE&chartGenre=KPOP&chartDate='
year = '2020#params%5Bidx%5D='
page = '1'
data = requests.get(url + year +page, headers = header)
soup = bs(data.text, 'html.parser')

soup.select('div#tb_list')

```

```python
album_img = soup.select('table tr td .wrap img')
titles = soup.select('table tr td .rank01 a')
artists = soup.select('table tr td .rank02 a')

img_url_list =[]
title_list = []

for i in range(100):
    img_url_list.append(album_img[i].get('src'))
    title_list.append(titles[i].text)
    
```

```python
for i in range(100):
    print((i+1),'위', title_list[i], img_url_list[i])
    
```

```python
# 가수 시도1
a = []
artist_list = []

artists = soup.select('table tr td .rank02')
for i in range(100):
    artist = artists[i].select('a')
    num = int(len(artist)/2)
    for j in range(num):
        a.append(artist[j].text)
    artist_list.append(a)
    a = []

print(artist_list)
for i in range(100):
    print(artist_list[i])
```
